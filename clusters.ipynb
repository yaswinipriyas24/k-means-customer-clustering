```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
x=data.iloc[:,[3,4]].values
print(x)
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title("The elbow point graph")
plt.xlabel("Number of clusters")
plt.ylabel('WSCC')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y=kmeans.fit_predict(x)
print(y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],c='green',label='Cluster 1')
plt.scatter(x[y==1,0],x[y==1,1],c='red',label='Cluster 2')
plt.scatter(x[y==2,0],x[y==2,1],c='yellow',label='Cluster 3')
plt.scatter(x[y==3,0],x[y==3,1],c='blue',label='Cluster 4')
plt.scatter(x[y==4,0],x[y==4,1],c='magenta',label='Cluster 5')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x[y == 0, 0], x[y == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(x[y == 1, 0], x[y == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(x[y == 2, 0], x[y == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(x[y == 3, 0], x[y == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(x[y == 4, 0], x[y == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids', marker='*')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
